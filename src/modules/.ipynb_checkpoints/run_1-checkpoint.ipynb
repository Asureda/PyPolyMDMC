```python
def write_data_to_vtk(points, filename, datasets, volume_normalization=True):
    """Creates a VTK object of the mesh

    Parameters
    ----------
    points : list or np.array
        List of (X,Y,Z) tuples.
    filename : str
        Name of the VTK file to write.
    datasets : dict
        Dictionary whose keys are the data labels
        and values are the data sets.
    volume_normalization : bool, optional
        Whether or not to normalize the data by
        the volume of the mesh elements.

    Raises
    ------
    ValueError
        When the size of a dataset doesn't match the number of mesh cells

    Returns
    -------
    vtk.vtkStructuredGrid
        the VTK object
    """

    import vtk
    from vtk.util import numpy_support as nps



    vtk_grid = vtk.vtkStructuredGrid()

    vtk_grid.SetDimensions(*[dim + 1 for dim in rc.tallies.mesh_filter.mesh.dimension])

    vtkPts = vtk.vtkPoints()
    vtkPts.SetData(nps.numpy_to_vtk(points, deep=True))
    vtk_grid.SetPoints(vtkPts)

    # create VTK arrays for each of
    # the data sets

    # maintain a list of the datasets as added
    # to the VTK arrays to ensure they persist
    # in memory until the file is written
    datasets_out = []
    for label, dataset in datasets.items():
        dataset = np.asarray(dataset).flatten()
        datasets_out.append(dataset)

        #if volume_normalization:
        #    dataset /= rc.tallies.mesh_filter.mesh.calc_mesh_volumes()

        dataset_array = vtk.vtkDoubleArray()
        dataset_array.SetName(label)
        dataset_array.SetArray(nps.numpy_to_vtk(dataset),
                       dataset.size,
                       True)
        vtk_grid.GetCellData().AddArray(dataset_array)

    # write the .vtk file
    writer = vtk.vtkStructuredGridWriter()
    writer.SetFileName(str(filename))
    writer.SetInputData(vtk_grid)
    writer.Write()

    return vtk_grid
```

```python
import numpy as np
from scipy.special import erf
from scipy.optimize import minimize,leastsq, curve_fit


def gauss(x):
    return np.exp( -0.5 * x**2 / np.sqrt( 2 * np.pi ) )


def Phi(x):
    return ( 0.5 * ( 1. + erf(x/np.sqrt(2) ) ) )


def skewed(x, x0, s, a):
    return 2./s * gauss( ( x - x0 ) / s ) * Phi( a * ( x - x0 ) / s)


def my_double_peak(x, A0, x0, s0, a, A1, x1, s1):
    return A0 * skewed( x, x0, s0, a ) + A1 / s1 * gauss( ( x - x1 ) / s1 )
```

```python
def my_peak(x, A0, x0, s0, a):
    return A0 * skewed( x, x0, s0, a )

def exponenial_func(x,x0, a, b, c):
    return a*np.exp(-b*(x-(x0)))+c
```

```python
import numpy as np
from scipy.integrate import solve_ivp


def hydrogen_diffusion(t, y, D, L, bc_left, bc_right):
    """
    Solves the one-dimensional diffusion equation with Dirichlet boundary conditions for hydrogen gas flowing in a metal:
    
    du/dt = D * d^2u/dx^2
    
    The boundary conditions are:
    u(0, t) = bc_left
    u(L, t) = bc_right
    
    Parameters
    ----------
    t: float
        The time at which to evaluate the solution.
    y: numpy array
        The concentration of hydrogen atoms at time t. It should have shape (N,) where N is the number of grid points.
    D: float
        The diffusion coefficient of hydrogen atoms in the metal.
    L: float
        The length of the metal sample.
    bc_left: float
        The concentration of hydrogen atoms at the left boundary (x = 0).
    bc_right: float
        The concentration of hydrogen atoms at the right boundary (x = L).
    
    Returns
    -------
    dy/dt: numpy array
        The time derivative of the concentration of hydrogen atoms at time t. It has the same shape as y.
    """
    N = y.shape[0]
    dy_dt = np.zeros_like(y)
    dy_dt[0] = bc_left  # Dirichlet boundary condition at x = 0
    dy_dt[-1] = bc_right  # Dirichlet boundary condition at x = L
    dy_dt[1:-1] = D * (y[2:] - 2 * y[1:-1] + y[:-2]) / (L / (N - 1))**2
    return dy_dt

L = 0.5  # Longitud del lecho metálico [cm]
dx = L / 1000  # Tamaño de la discretización [cm]
nx = int(L/dx)
C0 = 1.0  # Concentración inicial [mol/cm^3]
k1 = 0.1  # Tasa de recombinación [1/s]
k2 = 0.01  # Tasa de difusión intersticial [cm^2/s]


# Constante de Boltzmann [eV/K]
k_B = scipy.constants.physical_constants['Boltzmann constant in eV/K'][0]

# Masa del tritio [kg]
M = scipy.constants.physical_constants['triton mass'][0]

# Energía libre de Gibbs del tritio en el zirconio [eV]
ΔG = 2.2

# Temperatura [K]
T = 300

# Calculamos el coeficiente de difusión del tritio en el zirconio a partir de la energía libre de Gibbs
def diffusion_coefficient(T, ΔG, M):
  """Calcula el coeficiente de difusión a partir de la energía libre de Gibbs y la masa del difusivo.
  
  Args:
    T: Temperatura [K]
    ΔG: Energía libre de Gibbs [eV]
    M: Masa del difusivo [kg]
    
  Returns:
    Coeficiente de difusión [m^2/s]
  """
  return 1.0 / (6 * np.pi * M * k_B * T) * np.exp(-ΔG / (k_B * T))

D = diffusion_coefficient(T, ΔG, M)*10000  # Coeficiente de difusión del tritio en el zirconio [m^2/s]

bc_left = 1.0e24  # concentration of hydrogen atoms at x = 0 (atoms/m^3)
bc_right = 0.0  # concentration of hydrogen atoms at x = L (atoms/m^3)

# Discretize the domain
x = np.linspace(0, L, nx)  # grid points (m)

# Set the initial condition
t0 = 0.0  # initial time (s)
u0 = np.zeros_like(x)  # initial concentration of hydrogen atoms (atoms/m^3)
u0[0] = bc_left  # Dirichlet boundary condition at x = 0
u0[-1] = bc_right  # Dirichlet boundary condition at x = L

# Set the time steps for the solution
t_span = (t0, 10.0)  # start and end times (s)
dt = 0.005  # time step size (s)

# Solve the diffusion equation using the solve_ivp function
sol = solve_ivp(lambda t, y: hydrogen_diffusion(t, y, D, L, bc_left, bc_right), t_span, u0, t_eval=np.arange(t0, t_span[1], dt))
```
