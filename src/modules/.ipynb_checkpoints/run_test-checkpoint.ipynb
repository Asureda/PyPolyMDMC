```python
import numpy as np
import matplotlib.pyplot as plt

type1 = "NH"
type2 = "Cl"

nbins = 100
rmin = 0.0
rmax = r_max

id1 = ensemble.system.part.select(type=ensemble.type_particles[type1]).id
id2 = ensemble.system.part.select(type=ensemble.type_particles[type2]).id
obs.add_observable("rdf_nh_cl", obs.rdf_observable, id1, id2, r_min, r_max, r_bins)
obs.update_accumulators(ensemble.system)
ensemble.system.integrator.run(2)


dist = []
for pair in ensemble.system.part.pairs():
    if pair[0].type==ensemble.type_particles[type1] and pair[1].type==ensemble.type_particles[type2]:
        dist.append(ensemble.system.distance(pair[0], pair[1]))

n1=len(ensemble.system.part.select(type=ensemble.type_particles[type1]).id)
n2=len(ensemble.system.part.select(type=ensemble.type_particles[type2]).id)

rdf_nh_cl_observable_accumulator = obs.observables["rdf_nh_cl"]
radial = np.linspace(r_min,r_max,r_bins)
rdf_nh_cl = np.mean(rdf_nh_cl_observable_accumulator.time_series(),axis=0)


def rdf2(distances, nA, nB, box_size, bin_count=100):
    """Calcula la función de distribución radial (RDF) de un conjunto de distancias entre partículas.

    Args:
        distances (np.array): Array 1D con las distancias entre pares de partículas.
        particle_count (int): El número total de partículas en el sistema.
        box_size (float): El tamaño de la caja en unidades de longitud.
        bin_count (int, optional): El número de bins en los que se divide el rango de distancias. Por defecto es 100.

    Returns:
        Tuple: Un par de arrays numpy con las distancias radiales y la RDF correspondiente, normalizada por la cantidad de pares de partículas y el volumen de la caja.
    """
    n = len(dist)
    particle_count = nA*nB
    total_number = nA+nB
    bin_edges = np.linspace(0, box_size / 2, bin_count + 1)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    bin_volume = (4 / 3) * np.pi * (bin_edges[1:]**3 - bin_edges[:-1]**3)
    counts, _ = np.histogram(distances, bins=bin_edges)
    density = n / (box_size**3)
    rdf_values = counts / (bin_volume * density )
    return bin_centers, rdf_values

bin_centers,rdf2 = rdf2(dist, n1, n2, 2*r_max, bin_count=200)
# Graficamos g(r)
# Normalizamos el histograma de distancias para obtener g(r)
n = len(dist)
vol = (4 / 3) * np.pi *r_max**3
rho_bulk = n / vol

plt.plot(radial, rdf_nh_cl, label = "RDF espresso")
plt.scatter(bin_centers, rdf2,s=3,color="red",label = "RDF implementation")
plt.xlabel('Distancia radial')
plt.ylabel('Función de distribución radial')
plt.legend()
plt.show()

```

```python
def do_binning_analysis(data, bin_size):
    n_samples = len(data)
    n_bins = n_samples // bin_size
    bin_avgs = np.mean(data[:n_bins * bin_size].reshape((n_bins, -1)), axis=1)
    return np.std(bin_avgs, ddof=1.5) / np.sqrt(n_bins)
```

```python
sizes = np.arange(3, 6001, dtype=int)
sems = np.zeros(6001 - 3, dtype=float)
for s in range(len(sizes)):
    sems[s] = do_binning_analysis(particle_distance_time_data[:], sizes[s])

plt.figure(figsize=(10, 6))
plt.plot(sizes, sems, "x")
plt.xscale("log")
plt.xlabel("$N_B$")
plt.ylabel("SEM")
plt.show()
```

```python
from scipy.optimize import curve_fit

# only fit to the first couple of SEMs
CUTOFF = 5000

# sizes of the corresponding bins
sizes_subset = np.arange(3, 3 + CUTOFF, dtype=int)

def fit_fn(x, a, b, c):
    return -np.exp(-a * x) * b + c

fit_params, _ = curve_fit(fit_fn, sizes_subset, sems[:CUTOFF], (0.05, 1, 0.5))

fit_sems = fit_fn(sizes, *fit_params)

plt.figure(figsize=(10, 6))
plt.plot(sizes, sems, "x", label="binning analysis")
plt.plot(sizes, fit_sems, "-", label="fit")
plt.xscale("log")
plt.xlabel("$N_B$")
plt.ylabel("SEM")
plt.legend()
plt.show()

print(f"Final Standard Error of the Mean: {fit_params[2]:.4f}")
```
